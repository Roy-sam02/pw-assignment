# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.





Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations in a wide variety of formats and on various platforms. It provides a high-level interface for drawing attractive and informative statistical graphics. Matplotlib is widely used for data visualization tasks in fields such as data analysis, scientific research, machine learning, and more.

Matplotlib is used to create different types of plots and charts to visualize data and gain insights from it. Some common types of plots that can be created using the pyplot module of Matplotlib include:

Line Plot: A line plot is used to display data points connected by lines. It's often used to show trends or patterns in data over a continuous interval.

Bar Plot: Bar plots are used to compare the values of different categories by representing them as rectangular bars. They are effective for comparing categorical data.

Histogram: Histograms are used to represent the distribution of continuous data. They group data into bins and display the frequency or density of data points in each bin.

Scatter Plot: Scatter plots are used to visualize the relationship between two variables. Each data point is represented as a dot on the graph, with one variable on the x-axis and the other on the y-axis.

Pie Chart: A pie chart is used to represent the proportions of different categories as slices of a circle. It's useful for displaying parts of a whole or relative percentages.

## What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


A scatter plot is a type of data visualization that uses dots to represent individual data points on a two-dimensional (2D) plane. Each dot in a scatter plot corresponds to a single data point and is positioned based on the values of two variables: one plotted along the x-axis and the other plotted along the y-axis.

Scatter plots are commonly used to visualize the relationship between two continuous variables and to identify patterns, trends, and correlations between them. They are particularly useful for understanding the distribution of data points, the strength and direction of relationships, and the presence of any outliers or clusters within the data.

```python
import matplotlib.pyplot as pyt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
pyt.scatter(x,y)
pyt.xlabel("this is my x graph")
pyt.ylabel("this is my y graph")
pyt.title("this is my x vs y plot")
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.Use the following data:


The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. This function is particularly useful when you want to display multiple plots or charts in a single figure, each representing different aspects of your data or different visualizations. The subplot() function allows you to arrange these subplots in rows and columns, making it easier to compare and analyze different plots side by side.

```python
import matplotlib.pyplot as plt
import numpy as np
plt.subplot(2,2,1)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x,y)
plt.subplot(2,2,2)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x,y)
plt.subplot(2,2,3)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x,y)
plt.subplot(2,2,4)
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x,y)
```

A bar plot is a type of data visualization that uses rectangular bars to represent data values. Each bar corresponds to a category or group, and the length or height of the bar represents the value of a specific variable associated with that category. Bar plots are particularly useful for comparing the values of different categories, showing their relationships, and identifying patterns or trends in the data.

Bar plots are used for various purposes, such as:

Comparing Categories: Bar plots provide a clear visual representation of how the values of different categories compare to each other.

Categorical Data: Bar plots are ideal for visualizing categorical data, where each category represents a discrete group, such as different products, regions, or time periods.

Frequency Distribution: Bar plots can show the frequency or count of occurrences for each category.

Highlighting Differences: Bar plots make it easy to identify which categories have higher or lower values.

Presenting Data: Bar plots are often used in presentations and reports to provide an intuitive way to convey information to an audience.

```python
import matplotlib.pyplot as plt
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
```

# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.box1 = np.random.normal(100, 10, 200) box2 = np.random.normal(90, 20, 200)



A box plot, also known as a box-and-whisker plot, is a type of data visualization that provides a summary of the distribution of a dataset. It displays key statistical measures and identifies potential outliers in the data. Box plots are particularly useful for understanding the spread, central tendency, and variability of data across different categories or groups.

A box plot consists of several key components:

Box: The main body of the box plot represents the interquartile range (IQR), which contains the middle 50% of the data. The box spans from the first quartile (25th percentile) to the third quartile (75th percentile).

Median Line: A line (usually horizontal) inside the box represents the median value of the data, which is the middle value when the data is sorted.

Whiskers: Lines extending from the box (usually vertical) represent the data points beyond the quartiles. They typically extend to the minimum and maximum values within a certain range, which can also be influenced by the presence of outliers.

Outliers: Individual data points outside the whiskers are considered outliers. They are represented as individual points or dots on the plot.

Box plots are used for various purposes, including:

Visualizing Data Distribution: Box plots provide a clear visual representation of how data is distributed, including the spread and skewness.

Identifying Outliers: Outliers, which are data points significantly different from the rest of the data, are easily identified in a box plot.

Comparing Distributions: Box plots are useful for comparing the distribution of data between different categories or groups.

Understanding Central Tendency: The median and quartiles displayed in the box plot give insights into the central tendency and spread of the data.

Assessing Variability: Box plots help in assessing the variability of data and understanding whether the data is tightly clustered or widely spread.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.boxplot([box1,box2],labels=['Box 1','Box2'])
plt.xlabel('boxes')
plt.ylabel('values')
plt.title('box plot example')
plt.show()
```
