## 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for creating informative and attractive statistical graphics. Here are five common plots you can create using Seaborn, along with their typical use cases:

Histograms:

Use: To visualize the distribution of a single variable.
Description: Histograms display the frequency distribution of a continuous or discrete variable by dividing it into bins and showing the number of data points in each bin. They help you understand the data's central tendency and spread.
Scatter Plots:

Use: To explore the relationship between two continuous variables.
Description: Scatter plots represent individual data points as dots on a two-dimensional plane, with one variable on the x-axis and another on the y-axis. They are useful for identifying patterns, trends, and correlations between variables.
Bar Plots:

Use: To compare categories or discrete data.
Description: Bar plots display data as rectangular bars, with the height or length of each bar corresponding to the data value. They are suitable for showing comparisons between different categories or groups, such as comparing sales figures for different products.
Box Plots (Box-and-Whisker Plots):

Use: To visualize the distribution of a continuous variable and identify outliers.
Description: Box plots provide a summary of the distribution of a variable, including the median, quartiles, and potential outliers. They are helpful for detecting skewness and understanding the spread of the data.
Heatmaps:

Use: To visualize the relationships between two categorical variables using color intensity.
Description: Heatmaps are a way to represent data in a tabular format using colors. They are commonly used to visualize correlation matrices, showing how variables are related to each other. Darker colors indicate stronger relationships.
These are just a few examples of the types of plots you can create with Seaborn. Seaborn also offers many other plot types, such as line plots, violin plots, pair plots, and more, making it a versatile library for data visualization tasks. The choice of plot depends on the nature of your data and the specific insights you want to extract from it.

## 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns
import matplotlib.pyplot as plt
flight=sns.load_dataset("flights")
pivot_data=flight.pivot_table(index="month",columns="year",values="passengers")
plt.figure(figsize=(10,6))
sns.heatmap(pivot_data,annot=True,fmt='d')  # this fmt makes the number in human readable form
plt.title("passenger count by month and year")
plt.xlabel("year")
plt.ylabel("months")
plt.show()
```
