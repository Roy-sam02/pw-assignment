What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.


In Bokeh, glyphs are visual markers or shapes that you can add to a plot to represent your data points. Glyphs can be used to create various types of charts and visualizations, such as scatter plots, line plots, bar charts, and more. Bokeh provides a variety of glyph methods for adding different types of markers or shapes to your plots.

Some common glyph methods in Bokeh include:

circle: Adds circular markers to the plot.

square: Adds square markers to the plot.

line: Draws lines between data points.

patch: Fills areas defined by a set of coordinates.

rect: Adds rectangles to the plot.

segment: Draws line segments between pairs of points.

wedge: Adds wedges (partial circles) to the plot.

vbar and hbar: Creates vertical or horizontal bar charts.

How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?


```python
from bokeh.plotting import figure, show
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p = figure(title="Customized Bokeh Plot", background_fill_color="lightgray", border_fill_color="white")

# Additional customization options can include text_font, text_font_style, title_location, etc.
p.xaxis.axis_label = "X-axis Label"
p.yaxis.axis_label = "Y-axis Label"

# You can further customize ticks and tick labels, e.g., p.xaxis.major_label_text_font_size, p.yaxis.major_tick_line_color, etc.
p.legend.title = "Legend Title"
p.legend.label_text_font_size = "12pt"

# You can also customize other legend attributes like position, background_fill_color, border_line_color, etc.
p.line(x, y, line_color="blue", line_width=2, line_dash="dashed")
p.circle(x, y, size=10, fill_color="red", line_color="black")
p.xgrid.grid_line_color = "gray"
p.ygrid.grid_line_color = "gray"
p.xgrid.grid_line_dash = "dashed"
p.ygrid.grid_line_dash = "dotted"
p.title.text_font_style = "italic"
p.xaxis.axis_label_text_font_size = "14pt"
p.legend.label_text_font = "times"
p.background_fill_color = "lightgray"
p.border_fill_color = "white"
p.aspect_ratio = 1.0  # Set to 1.0 for a square plot
show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in
real time?

A Bokeh server is a feature of the Bokeh library that allows you to create interactive web applications with dynamic and real-time updating plots. Bokeh server enables you to build data-driven dashboards, applications, and visualizations that can respond to user interactions, streaming data, or scheduled updates.

Key features and benefits of Bokeh server include:

Real-Time Interactivity: Bokeh server applications can provide real-time interactivity by responding to user input events, such as clicking, hovering, or dragging, and updating the plot or other visual elements accordingly.

Streaming Data: You can create applications that continuously update with streaming data, enabling live data visualizations that change as new data arrives.

Interactive Widgets: Bokeh provides a wide range of interactive widgets (e.g., sliders, buttons, dropdowns) that can be added to your application to control plot parameters and trigger updates.

Cross-Filtering: You can build applications that allow users to filter and explore large datasets interactively by selecting data points or specifying filter criteria.

Custom Callbacks: Bokeh server allows you to define custom Python callbacks that execute when specific events occur. These callbacks can be used to update the plot or perform other actions based on user interactions or data changes.

Here's a basic outline of how you can use Bokeh server to create interactive plots that can be updated in real-time:

Define the Plot: Start by defining the initial plot and layout that you want to display in your application.

Create a Function: Write a Python function that updates the plot based on user interactions, streaming data, or other events. This function will be called whenever the plot needs to be updated.

Define Callbacks: Attach the Python function as a callback to specific events or widgets in your application. For example, you can attach the function to a button click or a slider value change.

Run the Bokeh Server: Use the Bokeh server command to launch your application, making it accessible via a web URL.

Access the Application: Users can access and interact with your application by visiting the provided URL. The application responds to their actions and updates the plot accordingly.

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?**bold text**

Install Flask (if you haven't already):
pip install Flask
Create a Flask App:

Create a Flask application by defining a Python script (e.g., app.py). Import the necessary modules and set up your Flask app:
from flask import Flask, render_template
from bokeh.embed import components
from your_bokeh_plot_script import create_bokeh_plot  # Replace with your Bokeh plot script

app = Flask(__name__)
Define a Route:

Define a route that will render an HTML template containing your Bokeh plot. In this route, you'll create the Bokeh plot and extract its components (div and script) using the components function:
@app.route('/')
def index():
    plot = create_bokeh_plot()  # Replace with how you create your Bokeh plot
    script, div = components(plot)
    return render_template('index.html', script=script, div=div)
Create an HTML Template:

Create an HTML template (e.g., templates/index.html) that will render your Bokeh plot using the script and div components provided by Bokeh:

<!DOCTYPE html>
<html>
<head>
    <title>Bokeh Plot Example</title>
    {{ script|safe }}
</head>
<body>
    <h1>My Bokeh Plot</h1>
    {{ div|safe }}
</body>
</html>
Run the Flask App:

In your app.py script, add the following to run the Flask app:
if __name__ == '__main__':
    app.run(debug=True)
Start the Flask App:

Run your Flask application:
python app.py
Access the Bokeh plot by visiting http://localhost:5000 in your web browser.
